```python
import pandas as pd
JA = pd.read_csv("Jamaican Creole.csv")
HA = pd.read_csv("hatian_creole_dictionary_v4.csv")
MC = pd.read_csv("martinique.csv")
LC = pd.read_csv("stLucia.csv")
SC = pd.read_csv("suriname.csv")
Lou = pd.read_csv("louisiana_creole_dictionary.csv")
```

```python
JA = JA[['Words','Definition','Country']]
JA.rename(columns={"Words": "creole_word", "Definition": "word", "Country":"creole name"}, inplace=True)
JA
```

```python
SC.rename(columns={"Word": "word", "Creole_word": "creole_word", "Creole_name":"creole name"}, inplace=True)
SC
```

```python
MC.rename(columns={"Word": "word", "Creole_word": "creole_word", "Creole_name":"creole name"}, inplace=True)
MC
```

```python
LC.rename(columns={"Word": "word", "Creole_word": "creole_word", "Creole_name":"creole name"}, inplace=True)
LC
```

```python
to_merge = [JA,HA,Lou,LC,MC,SC]

merged = pd.concat(to_merge)
merged.to_csv("Creole_merged_CSV.csv",index=False)
merged.to_json("Creole_merged_JSON.json",orient='records')
```
