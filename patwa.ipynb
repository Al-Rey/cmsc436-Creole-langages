```python
import re, pprint
from urllib import request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://jamaicanpatwah.com/dictionary"
response = request.urlopen(url).read().decode('utf8')
raw = BeautifulSoup(response, 'html.parser')
print(raw)
```

```python
setraw = raw.find_all('div', attrs={'class': 'dictionary-alphabet'})
dictionaryLinks = str(setraw)
dictionaryLinks = dictionaryLinks.split("</a>")
dictionaryLinks = [x.replace("</li>\n<li>\n<a href=\"",'') for x in dictionaryLinks]
dictionaryLinks = [x[:-3] for x in dictionaryLinks]
dictionaryLinks.pop(0)
dictionaryLinks.pop(-1)
```

```python
eachword = []
for x in dictionaryLinks:
    url = x
    response = request.urlopen(url).read().decode('utf8')
    raw = BeautifulSoup(response, 'html.parser')
    getletters = raw.find_all('div', attrs={'class': 'words'})
    getletters = str(getletters).replace("<div class=\"words\">\n<!-- Create cloumns-->\n<div class=\"col\">\n<ul>\n<li>\n<a href=",'')
    getletters = str(getletters).split("</a>")
    getletters  = [y.replace("</li>\n<li>\n<a href=\"",'').split("\">") for y in getletters ]
    getletters.pop(-1)
    eachword.append(getletters)
```

```python
wordlinks = []
removestuff = ["\n</li>\n</ul>\n</div>\n<div class=\"col last","\n</li>\n</ul>\n</div>\n<!-- end columns-->\n<br class=\"br\"/><br/>\n<!--paginations-->\n<div class=\"global-pagination","\n</li>\n<li class=\"cut","\n</li>\n</ul>\n</div>\n<div class=\"col"]
for i in range(0,len(eachword)):
    if eachword[i][-1][1] == "next »":
        for j in range(0,len(eachword[i])):
            if eachword[i][j][0] not in removestuff:
                wordlinks.append(eachword[i][j][0].replace("\xa0<a href=\"",'').replace('\n','').replace('[\"',''))

wordlinks
```
